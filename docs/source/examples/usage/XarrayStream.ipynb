
## Setup the Config

```python
from ioos_qc.config import Config

config = """
    streams:
        variable1:
            qartod:
                aggregate:
                gross_range_test:
                    suspect_span: [20, 30]
                    fail_span: [10, 40]
"""
c = Config(config)
c.config
```

## Setup the sample data

```python
import numpy as np
import xarray as xr
import pandas as pd

rows = 50
data_inputs = {
    'time': pd.date_range(start='01/01/2020', periods=rows, freq='D'),
    'z': 2.0,
    'lat': 36.1,
    'lon': -76.5,
    'variable1': np.arange(0, rows),
}
df = pd.DataFrame(data_inputs).set_index('time')
ds = xr.Dataset.from_dataframe(df)
ds
```

## Setup the XarrayStream

```python
from ioos_qc.streams import XarrayStream

xs = XarrayStream(ds)
xs
```

## Run the XarrayStream through the Config

```python
results = xs.run(c)
results
```
